#Function Application and Mapping

Series or DataFrame objects:

- Series .map method - Applies function to each element in Series
- DataFrame .apply method - Applies function column- or row-wise
- DataFrame .applymap method - Applies function to each element in DataFrame

```python
# Apply NumPy universal function
df = DataFrame(np.random.rand(3,5))
print(df)
np.exp(df)
```

```python
# Series map method - Numerical computation
ser = Series(np.random.rand(10))
print(ser)
ser.map(lambda x: x ** 2)
```

```python
ser = Series(np.random.rand(3))
print(ser)

def squared(x):
  return x**2

ser.map(squared)
```

```python
# Series .map method - String operation
Series({1: 'The Green Mile', 2: 'Black Panther', 3: 'The Bourne Identity'}).map(lambda s: len(s.split()))
```

```python
# DataFrame .apply method - down each column (axis=0), across each row (axis=1)
print(df)
print(df.sum(axis=1))
print(df.apply(lambda arr: arr / arr.sum(), axis=1))
```

```python
# DataFrame .applymap method
print(df)
print(df.sum(axis=1))
df.applymap(lambda x: x / df.sum().sum())
```

#Date Range

```python
# Create date range
dser = pd.Series(pd.date_range(start='1/1/2018', end='1/10/2018').astype(str))
dser
```

```python
# str.contains
dser.str.contains('\d{4}')
```

```python
print(dser)
```

```python
# str.extract - Single group
dser.str.extract('-(\d\d)$',expand=True) #if expand is true then output is dataframe, otherwise it will be an object
```

```python
pat = '-(?P<month>\d\d)-'
m = re.search(pat, dser[0])
print(dser[0])
print(m.group('month'))
print(m.group(0))
print(m.group(1))
```

```python
# str.extract - Multiple groups
#dser.str.extract?

dser.str.extract('((?P<year>\d\d\d\d)-(\d\d))-\d\d', expand=True)
```

```python
# str.findall

dser.str.findall('(\d+?)')
```

```python
# str.match

dser.str.match('\d\d')
```

```python
# str.replace

dser.str.replace('-0','-')
```

```python
dser.str.split('-|/').str.get(0).str.slice(2,4)
```

```python
df = pd.read_html('https://www.world-airport-codes.com/world-top-30-airports.html', header=0, index_col=0)[0]
df.head(3)
```

```python
# Extract city/metropolitan area
df.insert(1, 'Metro', df['Location'].str.split(', ').str[0])
df.head(3)
```

```python
# Extract city/metropolitan area
df.insert(2, 'State', df['Location'].str.extract(', (.+)', expand=False))
del df['Location']
df.head(3)
```

```python
# Split and expand IATA and ICAO airport codes
codes = df['Code (IATA/ICAO)'].str.split('/', expand=True).head()
codes.columns = ['IATA', 'ICAO']
codes.head()
```

```python
# Insert codes into DataFrame
df.insert(4, 'IATA', codes['IATA'])
df.insert(5, 'ICAO', codes['ICAO'])
del df['Code (IATA/ICAO)']
df.head()
```
